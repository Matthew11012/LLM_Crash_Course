```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("metrics.csv")

plt.plot(df["step"], df["train_loss"], label="train loss")
plt.plot(df["step"], df["val_loss"], label="val loss")
plt.legend()
plt.show()

plt.plot(df["step"], df["val_ppl"], label="val perplexity")
plt.legend()
plt.show()
```
